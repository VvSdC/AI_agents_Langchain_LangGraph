### Sentence transformers on Hugging face

HuggingFace sentence transformers is a python framework for state-of-the-art sentence, text and image embeddings. One of the embedding model is used in the HuggingFaceEmbeddings class. We have also added an alias for Sentence Transformer Embeddings for users who are more familiar with directly using that package.

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name = "all-MiniLM-L6-v2")
```

```python
text = "I am Vummethala Venkata Sri Datta Charan"
query_result = embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
doc_result = embeddings.embed_documents(["I am Vummethala Venkata Sri Datta Charan. I am 23 years old.","I graduated from IIITDM Jabalpur in 2024"])
doc_result[0]
```
